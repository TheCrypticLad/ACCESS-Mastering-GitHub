Today we are working on Python using Google Colab, we learnt about the following variables in Python.

**Strings**
These are usually letters which have to be enclosed in quotation marks: e.g  "Shammah", "Favour", "Nathan", "Abideen" etc. Strings can sometimes be combined with other variables like numbers using the **f string method**

**Floats**
These are decimal numbers type Variable e.g 29.99, 30.75, and so on and so forth.

**Integers**
These are basically just numbers. e.g 90, 100, 100000, 787627557.

**Boolean**
These are True/False Statemnts, in the entire data session they're the most tricky, cause in as much as just putting **true** in a sentence can trigger a **Boolean** command
